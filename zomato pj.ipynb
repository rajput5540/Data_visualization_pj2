Zomato data analysis project 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato data .csv")
```

```python
dataframe
```

convert the datatype of column - rate

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

TYPE OF RESTURANT

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('type of resturant')
```

Voting for resturant

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='red',marker="o")
plt.xlabel("Type of resturant",c="black",size=20)
plt.ylabel("Votes",c="black",size=20)
```

Rating of resturant

```python
plt.hist(dataframe['rate'],bins=8)
plt.title('rating distribution')
plt.show()
```

Average orders spending by couples

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

which mode receive maximum rating

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

which type of resturent receive more offline orders

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()
```
